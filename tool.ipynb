```python
import subprocess
from pathlib import Path

def compress_video(
    input_path,
    output_path=None,
    codec="libx265",   # 不兼容就改 'libx264'
    crf=26,            # 18 清晰但大，26~28 比较省空间
    preset="slow",     # medium / slow / slower
    audio_bitrate="96k"
):
    """
    压缩视频（不改变 fps），尽量减小体积 & 控制质量损失。

    :param input_path: 输入视频路径
    :param output_path: 输出视频路径（默认同目录加 _compressed）
    :param codec: 视频编码器: 'libx265' 或 'libx264'
    :param crf: 质量参数，越小越清晰、文件越大（推荐 23~30）
    :param preset: 编码速度/压缩比，越慢体积越小
    :param audio_bitrate: 音频码率
    """
    input_path = Path(input_path)
    if output_path is None:
        output_path = input_path.with_name(input_path.stem + "_compressed.mp4")
    else:
        output_path = Path(output_path)

    cmd = [
        "ffmpeg",
        "-i", str(input_path),
        "-vcodec", codec,
        "-crf", str(crf),
        "-preset", preset,
        "-acodec", "aac",
        "-b:a", audio_bitrate,
        "-movflags", "+faststart",  # 提前写 moov，网页播放更快开始
        "-y",                       # 覆盖输出文件
        str(output_path)
    ]

    print("Running:", " ".join(cmd))
    subprocess.run(cmd, check=True)
    print(f"✅ 压缩完成：{output_path}")

if __name__ == "__main__":
    # 示例：直接改成你自己视频路径跑就行
    compress_video("WorldWander/static/videos/comparison/three2one/case5/TrajectoryCrafter.mp4")
```
